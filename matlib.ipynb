#    SCATTERPLOT

```python
a=np.array([1,2,3,4,5,6,7])
b=np.array([10,11,12,13,14,15,16])
plt.scatter(a,b) # scatterly spread the plot 
plt.show()
```

```python
a=np.array([1,4,6,8,9])
b=np.array([2,4,6,8,9])
c=np.array([5,8,9,2,5])
d=np.array([2,3,4,5,7])
plt.scatter(a,b,c,d) # used to compare between these plot 
```

```python
# color change in each data point 
a=np.array([1,4,6,8,9])
b=np.array([12,14,16,18,19])
c=np.array(["red","green","yellow","blue","black"])  # ye color bta rha h 
plt.scatter(a,b,c=c)
```

```python
a=np.array([1,4,6,8,9])
b=np.array([21,14,16,18,19])
c=np.array(["red","green","yellow","blue","black"])
size=np.array([100,200,300,400,500]) # ye size bta rha h each data point ki 
plt.scatter(a,b,s=size,alpha=0.5,c=c) # alpha=intensity of colour bta rha hn
plt.show()
```

```python
a=np.array([1,4,6,8,9])
b=np.array([21,14,16,18,19])
c=np.array(["red","green","yellow","blue","black"])
size=np.array([100,200,300,400,500])
plt.scatter(a,b,s=size,alpha=0.9,c=c)
plt.show()
```

```python
a=np.array([1,4,6,8,9])
b=np.array([21,14,16,18,19])
c=np.array(["red","green","yellow","blue","black"])
size=np.array([100,200,300,400,500])
plt.scatter(a,b,s=size,alpha=0.5,c=c,cmap="nipy_spectral") # cmap= used to make side color spectrum 
plt.colorbar()
plt.show()
```

#   BARPLOT

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y) # used to make a bar plot between catgeroical and numerical data
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y) # used to make a bar plot between catgeroical and numerical data(horizontal bar)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color="red") # used to make bar colored 
plt.show()
```

```python
# reduce and maximize sizeof bar
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,width=0.1) # width used to reduce the size of bar
plt.show()
```

#   PIECHART

```python
x=np.array([1,2,3,4])
mylabels=(["apple","banana","mango","orange"])
plt.pie(x,labels=mylabels)
plt.show()
```

```python
x=np.array([1,2,3,4])
mylabels=(["apple","banana","mango","orange"])
plt.pie(x,labels=mylabels,startangle=90)
plt.show()
```

```python
x=np.array([1,2,3,4])
mylabels=(["apple","banana","mango","orange"])
myexplode=[0,0.2,0,0]
plt.pie(x,labels=mylabels,explode=myexplode)
plt.show()
```

```python
x=np.array([1,2,3,4])
mylabels=(["apple","banana","mango","orange"])
myexplode=[0,0.2,0,0]
plt.pie(x,labels=mylabels,explode=myexplode,shadow="True")
plt.show()
```

```python
x=np.array([1,2,3,4])
mylabels=(["apple","banana","mango","orange"])
myexplode=[0,0.2,0,0]
mycolor=["black","white","black","white"]
plt.pie(x,labels=mylabels,explode=myexplode,colors=mycolor,shadow=True)
plt.show()
```

```python
x=np.array([1,2,3,4])
mylabels=(["apple","banana","mango","orange"])
myexplode=[0,0.2,0,0]
plt.pie(x,labels=mylabels,explode=myexplode,autopct="%1.1f%%") # autopct means % btana 
plt.show()
```

```python
x=np.array([1,2,3,4])
mylabels=(["apple","banana","mango","orange"])
myexplode=[0,0.2,0,0]
plt.pie(x,labels=mylabels,explode=myexplode)  # explode means upper part show uthana
plt.legend(title="colorss") # lengend m title show krna
plt.show()
```
